```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def quantize(float_value, min_range, max_range, zero=0):
    b = 8
    S = (max_range - min_range) / (2 ** b - 1)
    quantized = (float_value / S).astype(np.int8)
    return quantized


def to_float(uint_values, min_range, max_range):
    b = 8
    S = (max_range - min_range) / (2 ** b - 1)
    float_values = uint_values * S
    return float_values
```

```python
def display_error_histogram(values, min, max):
    quantized = quantize(values, min, max)
    float_values = to_float(quantized, min, max)
    error = float_values - values
    plt.hist(error, bins=100)
    plt.show()
    

def quantization_error_stats(values, min, max):
    quantized = quantize(values, min, max)
    float_values = to_float(quantized, min, max)
    error = float_values - values
    mean_error = np.mean(error)
    std_error = np.std(error)
    return mean_error, std_error
```

```python
values = np.random.uniform(0, 1, 1000)
min_range = 0
max_range = 1
display_error_histogram(values, min_range, max_range)
mean_error, std_error = quantization_error_stats(values, min_range, max_range)
print(mean_error, std_error)
```

```python
# Pour conclure, on peut dire que plus l'intervalle de quantification est petit, plus l'erreur est grande.
# Cela est dû au fait que l'erreur est proportionnelle à la taille de l'intervalle de quantification.
# Cependant, la moyenne et la variance de l'erreur restent quasiement constantes quelque soit l'intervalle de quantification.
```
